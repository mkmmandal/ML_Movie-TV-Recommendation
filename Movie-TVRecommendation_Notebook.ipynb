###  <font color=darkblue>  Here I am fetching the data for my application from TMDB website using its api.</font>
> I am using 2 api's to get the list of current popular movies and tv on TMDB. These lists are updates daily. These lists will become good enough starting base of movies for users to search and get recommendations.
>- Purpose of getting the data from the api and not from some csv/data file so that we can have `Real Time data.`
>- Also we can use different website apis later like 'IMDB', 'Rotten Tomatoes'.(To access those Api's we need subscription)

```python
# Get response of the api which will be json and make dataframe out of it.
# Apis have page input with values rangng from 1-1000 and each page has 20 movies/tv-series in its result.
# Iterate over 250 times with movie api to get to 250 page of data which will be 5000 movies data. Repeat same step for tv-series.


def getMovieData():
    response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=9f180c23e28c2109c30804fc9d8f367e&language=en-US&page=1')
    movieData=pd.DataFrame(response.json()['results'])
    for i in range(2,251):
        response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=9f180c23e28c2109c30804fc9d8f367e&language=en-US&page='+str(i))
        movieDatai=pd.DataFrame(response.json()['results'])
        movieData=pd.concat([movieDatai,movieData])
    return movieData

def getTvData():
    response=requests.get('https://api.themoviedb.org/3/tv/popular?api_key=9f180c23e28c2109c30804fc9d8f367e&language=en-US&page=1')
    tvData=pd.DataFrame(response.json()['results'])
    for i in range(2,251):
        response=requests.get('https://api.themoviedb.org/3/tv/popular?api_key=9f180c23e28c2109c30804fc9d8f367e&language=en-US&page='+str(i))
        tvDatai=pd.DataFrame(response.json()['results'])
        tvData=pd.concat([tvDatai,tvData])
    return tvData
    
```

```python
#making tag a new column which is basically bag of words made with all the above words list(language,genres,overview,cast,crew)
finalData['tags']=finalData.original_language+finalData.overview+finalData.cast1+finalData.crew+finalData.fgenre
```

```python
#this is our tag data in which we have title and tags
tagData=finalData.drop(['poster_path','cast1','crew','fgenre','overview','original_language'],axis=1)
```

```python
tagData
```

```python
tagData.tags=tagData.tags.apply(lambda x:" ".join(x)) #joining words from the list
```

```python
tagData.tags=tagData.tags.apply(lambda x:x.lower())# converting to lower case
```

```python
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
#This method is for doing stemming
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
tagData.tags=tagData.tags.apply(stem)
```

```python
tagData
```

## Model Deployment

#### Cosine Similarity

```python
#We will make vectors out of our tag data
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')#considering 5000 words and ignoring stop words
```

```python
vectors=cv.fit_transform(tagData.tags).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Deploying method of Cosine similarity
similarity=cosine_similarity(vectors)
```

```python
tagData.reset_index(inplace=True)
```

```python
#this method wil fetch the top near by vectors(movies title) from the similarity matrix of the chosen content
def recommend(content):
    recoList=[]
    content_index=tagData[tagData['title']==content].index[0]
    distances=similarity[content_index]
    content_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    for i in content_list:
        
        recoList.append(tagData.loc[i[0]].title)
    return recoList
```

```python
#this method will web scrap by hitting the url and give us ratings of all the reccos
def getRatings(imdbid):
    response=requests.get('https://www.imdb.com/title/'+imdbid+'/ratings/?ref_=tt_ov_rt')
    htmlr=response.content
    soup = BeautifulSoup(htmlr,'lxml')
    soup_result = soup.find_all("div",{"class":"allText"})
    rating=0.0
    for i in soup_result:
        a=i.get_text()
        rating=(a[80:100])
        break
    return rating
```
